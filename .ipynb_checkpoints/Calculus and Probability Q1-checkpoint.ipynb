In this question we are given $X_{1},... X_{n}$ i.i.d $U[0,1]$ r.v.

We will define $Y = max(X_{1},... X_{n}$)

We will calculate $\mathbb E{[Y]}$ and $Var[Y]$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf_y(t):
    return n * (t ** (n - 1))
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

x = np.linspace(0, 1, 10000)

for n in range(1, 11):
    plt.plot(x, pdf_y(x))

plt.show()
```

```python
def mean_y(n):
    return 1 - 2 / (n + 1)

def var_y(n):
    return n / ((n + 1)**2 * (n + 2))

x = np.linspace(1, 100, 100)

plt.plot(x, mean_y(x), color="r", label="calculated mean of Y")
plt.plot(x, var_y(x), color="g", label="calculated variance of Y")

plt.xlabel("number of U[0,1]")

plt.show()
```

```python
def heaviside(x):
    """See http://stackoverflow.com/a/15122658/554319"""
    y = 0.5 * (np.sign(x) + 1)
    y[:-1][np.diff(y) >= 0.5] = np.nan
    return y

x = np.linspace(-1, 1)

plt.plot(x, heaviside(x))
plt.ylim(-1, 2)
```
